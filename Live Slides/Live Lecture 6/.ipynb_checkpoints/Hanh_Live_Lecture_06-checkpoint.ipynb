#### Step 13.c. Check the item price type

```python
chipo.info()
chipo.head(5)
```

```python
type(chipo.item_price.iloc[0])
```

```python
# convert tu str -> float for item_price
chipo.item_price.iloc[0]
```

```python
type(float('2.39'))
```

```python
'$2.39'[1:]
```

```python
def convert_type1(x):
    '''
    convert string type to float for values
    
    params:
    
    '''
    return float(x.replace('$','')) # regex -> import re
```

```python
def convert_type(x):
    '''
    convert string type to float for values
    
    params:
    
    '''
    return float(x[1:])
```

```python
chipo['item_price_float'] = chipo.item_price.apply(convert_type)
```

```python
chipo.head()
```

```python
chipo.info()
```

```python


# sach: https://bedford-computing.co.uk/learning/wp-content/uploads/2015/10/Python-for-Data-Analysis.pdf
```

```python
#Anonymous function
convert_type2 = lambda x: float(x.replace('$','')) #busuness logic
chipo['item_price_float'] = chipo.item_price.apply(convert_type2)
chipo.head()
```

```python
type(chipo.item_price.iloc[0])
```

## II. Filtering & Sorting

### Step 18. How many products cost more than $10.00?

```python
chipo.head()
```

```python
chipo['price_per_item'] = chipo ['item_price_float'] / chipo['quantity']
chipo.head()

# chipo[chipo['item_price_float'] > 10]
```

```python
chipo['price_per_item'] = [x/y for x,y in list(zip ( chipo['item_price_float'], chipo ['quantity']))]
chipo.head()
```

```python
chipo[chipo['item_price_float'] > 10]
```

### Step 19. What is the price of each item? 
###### Print a data frame with only two columns item_name and item_price

```python
sample_chipo = chipo[['item_name', 'price_per_item']].drop_duplicates()
sample_chipo[sample_chipo['item_name'] == 'Chicken Bowl']
```

```python
chipo.groupby(['item_name', 'choice_description']).price_per_item.mean().to_frame().reset_index()
```
