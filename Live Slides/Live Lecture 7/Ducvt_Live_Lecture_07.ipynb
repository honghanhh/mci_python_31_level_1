### 2.1. Line graph

#### 2.1.1 Plot 1 line graph

```python
x = np.linspace(-10,10,100)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.title('Sin Function')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.grid()
plt.show()
```

```python
sin_df = pd.DataFrame({'x': x, 'y':y})
sin_df.head(5)
```

```python
sns.lineplot(data = sin_df, x='x', y='y') #DataFrame, Series
plt.show()
```

```python
x = np.linspace(-10,10,100)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.plot(x, sin, label='Sin')
plt.plot(x, cos, label='Cos')
plt.title('Function')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```

Method 2

```python
ax = plt.axes()
ax.plot(x,sin, label='Sin')
ax.plot(x,cos, label='Cos')
ax.set_title('Function')
ax.set_xlabel('X value')
ax.set_ylabel('Function of x')
#ax.set(xlabel='X value', ylabel='Function of x', title='Function')
ax.legend(loc='best')
plt.show()
```

#### 2.1.2 Plot 2 line graph

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1) #number of row, number of column, current position
plt.plot(x, sin)
plt.title('Sin Function')
plt.xlabel('X value')
plt.ylabel('Y value')

plt.subplot(1,2,2)
plt.plot(x, cos, c='red')
plt.title('Sin Function')
plt.xlabel('X value')
plt.ylabel('Y value')

plt.show()
```

```python
plt.figure(figsize=(15,5))

ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax1.plot(x, sin)
ax2.plot(x, cos)

ax1.set(xlabel='X', ylabel='Sin Function')
ax2.set(xlabel='X', ylabel='Cos Function')

plt.show()
```

### 2.3. Scatter plot

```python
iris.head(5)
```

```python
plt.scatter(iris.sepal_length, iris.sepal_width, label='sepal')
plt.scatter(iris.petal_length, iris.petal_width, label='petal')
plt.legend()
plt.show()
```

### 2.5. Pie chart

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
sex = tips.groupby('sex').total_bill.sum()
day = tips.groupby('day').total_bill.sum()
```

```python
#suplots
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].pie(sex, labels=sex.index, autopct='%.2f%%')
ax[1].pie(day, labels=day.index, autopct='%.2f%%');
```

```python
male =  tips[tips['sex']=='Male'].groupby('day').total_bill.sum()
male
plt.pie(male, labels=male.index, autopct='%.2f%%', explode=[0,0.3,0,0]);
```

```python
day = tips.groupby('day').total_bill.sum()
```
