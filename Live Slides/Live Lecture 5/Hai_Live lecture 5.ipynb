- Select all rows that sex is Male and total_ bill is larger than 10
- Select all rows that smoker is No and size is larger than 3
- Select all rows that sex is Female and day of weekends and time is Dinner

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
df3 = sns.load_dataset('tips')
df3
```

```python
condition = (df3['sex'] == 'Male') & (df3['total_bill'] > 10)
df3[condition]
```

```python
condition = (df3['smoker'] == 'No') & (df3['size'] > 3)
df3[condition]
```

```python
condition = (df3['sex'] == 'Female') & (df3['time'] == 'Dinner') & ((df3['day'] == 'Sun')|(df3['day'] == 'Sat')|(df3['day'] == 'Fri'))
df3[condition]
```

```python
con1 = (df3['sex'] == 'Female') & (df3['time'] ==' Dinner')
con2 = (df3['day'] == 'Fri') | (df3['day'] == 'Sat') | (df3['day'] == 'Sun')
condition = con1 & con2
df3[con1]
```

```python
left = df3.head(10)[['total_bill', 'tip']].reset_index()
left
```
