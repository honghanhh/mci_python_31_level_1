```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

x - creat an array range from -10 to 10 including 100 items

y - sin(x)

```python
x= np.linspace(-10, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.title('Sin Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

**Plot sin and cos function given x**

```python
x = np.linspace(-10, 10, 100)
sin = np.sin(x)
cos = np.cos(x)
plt.plot(x, sin, label='sin', c='green', linestyle='--')
plt.plot(x, cos, label='cos', c='red', linestyle='-.')
plt.legend(loc='best')
plt.show()
```

```python
ax = plt.axes()
ax.plot(x,sin)
ax.plot(x,cos)
plt.show()
```
