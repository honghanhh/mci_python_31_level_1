### 2.1. Line graph

```python
# We talk alittle bit about numpy 
# creat and array from -10 10 including 100 items and make y = sin(x)
```

```python
import numpy as np 
```

```python
x = np.linspace(-10,10,100) #inclusivestart and inclusivestop
```

```python
print(x)
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.title('Sin function')
plt.xlabel('x_value')
plt.ylabel('y_value')
plt.grid('major')
plt.show()
```

**plot cos and sin function given x**

```python
x= np.linspace(-10,10,100)
sin = np.sin(x)
cos = np.cos(x)
```

**Method1**

```python
plt.plot(x,sin,label = 'Sin',c ="red",linestyle='--')
plt.plot(x,cos,label = "Cos",c ='green')
plt.title("Cos and Sin fuction given X")
plt.xlabel('X_value')
plt.ylabel('Fuction of X')
plt.legend(loc='best')
plt.show()
#YOu have to check all the fuction again of syntax it is very important at al which you can learn easily
#if you wanna take more configuration of linstype you can consider it at: 
```

**It means that we add a frame's diagram by using plt**

```python
ax = plt.axes()
ax.plot(x,sin,label = 'Sin',c='red')
ax.plot(x,cos,label = 'Cos')
ax.set_title('Fucction')
ax.set_xlabel('X_value')
ax.set_ylabel('Y_value')
ax.legend(loc='best')
plt.show()
```
