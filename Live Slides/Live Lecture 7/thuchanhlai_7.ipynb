```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Tao 1 mang tu -10 den 10 va 100 phan tu
x = np.linspace(-10, 10, 100)
y= np.sin(x)
#Bieu dien 
plt.plot(x,y) # bieu dien theo bieu do, muon xoa phan mo ta DL, nhap ";" hoac plt.show()
plt.title('Bieu do Sin')
plt.xlabel('truc x')
plt.ylabel('truc y')
plt.grid()  # o Luoi

plt.show()
```

```python
# bieu dien do thi SD sns

sin_df=pd.DataFrame({'x':x, 'y':y})
sin_df
sin_df.head()
```

```python
## Plot sin and cos function given x
sns.lineplot(data=sin_df, x='x', y='y')
plt.show()
```

```python
## Plot sin and cos functinon given
x = np.linspace(-10, 10, 100)
sin = np.sin(x)
cos = np.cos(x)
# Mo ta thong tin bieu do
plt.plot(x, sin, label='sin', c= 'green', linestyle = '--' )
plt.plot(x, cos, label='cos' )
plt.title('Function')
plt.grid()
plt.xlabel('x value')
plt.ylabel('Function of x')
plt.legend(loc='best')

plt.show()
```

```python
# Method 2: sd axes
ax=plt.axes()
ax.plot( x, sin, label='sin' )
ax.plot( x, cos, label = 'cos')
# 
ax.set(xlabel = 'x value', ylabel = 'Function of x', title='Function' )
ax.legend(loc='best')
plt.show()

```

```python
#### 2.1.2 Plot 2 line graph
# subplot and subplots # chia thanh 2 bieu do

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(x, sin)
plt.title('sin Function')
plt.xlabel('x value')
plt.ylabel('y value')

# Bieu do thu 2, chi can thay doi thu tu cua # subplot

plt.subplot(1,2,2)
plt.plot(x, cos, c='red')
plt.title('cos Function')
plt.xlabel('x value')
plt.ylabel('y value')

plt.show()
```

```python
# cach 2 de khai bao
plt.figure(figsize=(15,5))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax1.plot(x,sin)
ax2.plot(x,cos, c='red')

ax1.set(xlabel='x value', ylabel='y value')
ax2.set(xlabel='x value', ylabel='y value')

plt.show()
```

```python
#SD # subplots
fig,ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(x, sin)
ax[1].plot(x, cos, c='red')

plt.show()
```

```python
### 2.5. Pie chart
# Bieu do tron 
# chi plt moi co, sns khong co

```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips['sex'].value_counts()
```

```python
#Tinh % so luong nam/nu trong tap DL
sex = tips.groupby('sex').total_bill.sum()
sex
```

```python
plt.pie(sex,labels=sex, autopct = '%.2f%%' );
```

```python
plt.pie(sex,labels=sex.index, autopct = '%.2f%%' );
```

```python
sex = tips.groupby('sex').total_bill.sum()
day = tips.groupby('day').total_bill.sum()
sex
day
```
