### 2.7 Join / Merge

- `Join` / `Merge`

````python
left = df3.head(10)[['total_bill', 'tip']].reset_index()
left
````

````python
right = df3.loc[5:15,'total_bill':'smoker'].reset_index()
right
````

````python
final = pd.merge(left, right, how='inner', on='index')
````

````python
final
````

````python
final2 = pd.merge(left, right, how='inner', on=['index','total_bill','tip'])
final2
````

````python
final = pd.merge(left, right, how='outer', on='index')
final
````

````python
final = pd.merge(left, right, how='left', on='index')
final
````

````python
final = pd.merge(left, right, how='right', on='index')
final
````

### 2.8 Group by

- `Group by`: `split` - `apply` - `combine`

Check average bill value per sex gender

- Key: `df3['sex']`
- Value: `df3['total_bill']`
- Expected out put: Male = ? , Female = ?

````python
df3.groupby('sex')['total_bill'].mean()
````

````python
df3.groupby('sex')['total_bill'].max()
````

````python
df3.groupby('sex')['total_bill'].min()
````

````python
df3.groupby('sex').agg({'total_bill':['mean','min','max']}) # .agg stand for aggregated
````

````python
df3.groupby('sex').agg({'total_bill':['mean','min','max'], 'tip':['mean','min','max']})
````
