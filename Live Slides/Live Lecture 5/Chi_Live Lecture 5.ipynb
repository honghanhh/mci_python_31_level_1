### 2.7. Join/Merge

```python
df3.head(5)
```

- join/merge

```python
left = df3.head(10)[['total_bill','tip']].reset_index()
left
```

```python
right = df3.loc[5:15,'total_bill':'smoker'].reset_index()
right
```

```python
final = pd.merge(left,right,how='inner',on=['index','total_bill','tip'])
final
```

### 2.8. Group by

- `Group by`:
    - `Split`:
    - `Apply`:
    - `Combine`:

Check average bill each sex has to pay in tips data

- Key: `df3['sex']`
- Value: `df3.['total_bill']`
- Expected output: female-?, male-?

```python
df3.groupby('sex').total_bill.mean()
#df3.groupby('sex')['total_bill'].mean()
```

```python
df3.groupby('sex').total_bill.median()
```

```python
df3.groupby('sex').total_bill.min()
```

```python
df3.groupby('sex').agg({'total_bill':['max','min','mean'],'tip':['max','min','mean']}) #aggregate
```
