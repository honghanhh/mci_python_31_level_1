```python
from collections import Counter
def detect_outlier(df,n,features):
    '''
    Input: dataframe and features
    OUtput:
    '''
    outliner_indices = []
    for col in features:
    #Step1: interare iver fearyres(columns)
        #Step2: Calcutae the 1st quantile (25%)
        Q1 = np.percentile(df[col], 25)
        #Step3: Calcutae the 1st quantile (75%)
        Q3 = np.percentile(df[col], 75)
        #Step4: Calcutae the IQR= Q3-Q1
        IQR= Q3-Q1
        #Step5:Calulate the outliner step
        outlier_step = 1.5*IQR
        #Step6: Index correponding with outlier value 
        outliner_list_col= df[(df[col]< Q1-1.5*IQR)|(df[col]>1.5*IQR)].index
        outliner_indices.extend(outliner_list_col)
    #Seleect rows contatting more than n outliners
    outliner_indices = Counter(outliner_indices)
    multiple_outliners = (k for k,v in outliner_indices.items() if v>n) #Taken k given v>n
    
    return multiple_outliners
```
