## 2. VISUALIZATION 

### 2.1. Line graph

#### 2.1.1 Plot 1 line graph

```
x - create an array range from -10 to 10 including 100 items
y - sin(x)
```

````python
x = np.linspace(-10,10,100)
x
````

````python
sin = np.sin(x)
sin
````

````python
x = np.linspace(-10,10,100)
cos = np.cos(x)
cos
````

#### a. Method 1

***Vẽ đồ thị hình sin***

````python
plt.plot(x,sin) #loại đồ thì mặc định là line
plt.title('Sin Function')
plt.xlabel('X value')
plt.ylabel('Sin of x')
plt.show()
````

````python
sin_df = pd.DataFrame({'x':x, 'y':sin})
sin_df.head()
````

````python
# Using Seaborn
sns.lineplot(data = sin_df, x = 'x', y = 'y'); #DataFrame, Series
# Using ';' at the end of the clause so you don't have to add plt.show() to hide the object of the plot
````

***Vẽ đồ thị hình cos***

````python
plt.plot(x, sin, label = 'sin', c='green', linestyle=':')
plt.plot(x, cos, label = 'cos', c='#cd282f', linestyle=(0,(5,1)))
plt.title('Sin & Cos X plot')
plt.xlabel('X value')
plt.ylabel('Function of X')
plt.legend(loc = 'best')
plt.show()
````

#### b. Method 2

````python
ax = plt.axes()
ax.plot(x,sin, label = 'Sin')
ax.plot(x,cos, label = 'Cos')
# ax.set_title('Function')
# ax.set_xlabel('X value')
# ax.set_ylabel('Function of X')
ax.set(xlabel='X value', ylabel='Function of X', title = 'Title using method 2')
ax.legend(loc='best')
plt.show()
````

#### 2.1.2 Plot 2 line graph

**Subplot** and **Subplots**

***Subplot***

````python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1) # 1 row, 2 columns, position 1
plt.title('Sin Function')
plt.plot(x,sin, c= 'red')
plt.xlabel('X value')
plt.ylabel('Sin city')
plt.subplot(1,2,2) # 1 row, 2 columns, position 2
plt.title('Cos Function')
plt.plot(x,cos)
plt.xlabel('X value')
plt.ylabel('Cos of X')
plt.grid()
plt.show()
````

````python
plt.figure(figsize=(15,5))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
ax1.plot(x,sin)
ax2.plot(x,cos, c='orange')
ax1.set(xlabel = 'x', ylabel='Sin(x)')
ax2.set(xlabel = 'x', ylabel = 'Cos(x)')
plt.show()
````

***Subplots***

````python
fig , ax = plt.subplots(1,2, figsize = (15,5))
ax[0].plot(x,sin)
ax[1].plot(x,cos, c = 'orange')
plt.show()
````

### 2.5. Pie chart

````python
tips = sns.load_dataset('tips')
tips.head(5)
````

````python
tips['sex'].value_counts()
````

````python
sex = tips.groupby('sex').total_bill.sum()
sex
````

````python
day = tips.groupby('day').total_bill.sum()
day
````

````python
plt.pie(sex, labels = sex.index, autopct='%.2f%%');
````

````python
sex1 = sex.reset_index()
sex1
````

````python
sns.barplot(data=sex1, x='sex', y='total_bill');
````

````python
#plt.figure(figsize = (15,5))
fig, ax = plt.subplots(1,2, figsize = (15,5))
ax[0].pie(sex, labels = sex.index, autopct = '%.2f%%')
ax[1].pie(day, labels = day.index, autopct = '%.f%%');
````

**Excercise**: pie chart visualizing total consumption (total_bill) of Male regarding days (day)

````python
menday = tips[(tips['sex'] == 'Male')].groupby('day').total_bill.sum()
menday
````

````python
menday2 = tips[(tips['sex'] == 'Male')].groupby('day').sum('total_bill')
menday2
````

````python
plt.pie(menday2(['day','tip']), labels=menday2['day'], autopct = '%.2f%%');
````

````python
plt.pie(menday, labels=menday.index, autopct = '%.2f%%');
````

````python
plt.pie(menday, labels=menday.index, autopct = '%.2f%%', explode=[0,0.1,0,0]);
````

### 2.6. Swarmplot

````python
sns.swarmplot(data = tips, x = 'time', y = 'total_bill', hue = 'smoker');
````

### 2.7. Catplot

Cat = Category Plot

````python
sns.catplot(data = tips, x = 'time', y = 'total_bill', hue = 'smoker', kind = 'swarm');
````

````python
sns.catplot(data = tips, x = 'time', y = 'total_bill', hue = 'smoker', kind = 'violin');
````

````python
# adding more conditions
sns.catplot(data = tips, x = 'time', y = 'total_bill', col = 'sex', hue = 'smoker', kind = 'violin');
````

````python
sns.catplot(data = tips, x = 'time', y = 'total_bill', row = 'sex', hue = 'smoker', kind = 'violin');
````

````python
sns.catplot(data = tips, x = 'time', y = 'total_bill', col = 'sex' , row = 'smoker', hue = 'smoker', kind = 'violin');
````
