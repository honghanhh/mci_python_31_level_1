*plot sin and cos function given x*

```python
x = np.linspace(-10,10,100)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.plot(x,sin,label = 'Sin',linestyle = '--')
plt.plot(x,cos,label = 'Cos',c = 'red',linestyle = 'dotted')
plt.title('function')
plt.xlabel('X value')
plt.ylabel('Function of X')
plt.legend(loc = 'best')
plt.show()
```

```python
ax = plt.axes()
ax.plot(x,sin,label = 'sin')
ax.plot(x,cos,label = 'cos')
#ax.set_title('Function')
#ax.set_xlabel('x value')
#ax.set_ylabel('function of x')
ax.set(xlabel = 'x value',ylabel = 'function of x',title = 'function')
ax.legend()
plt.show()
```

#### 2.1.2 Plot 2 line graph

***subplot and subplots***

```python
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
# x and sin(x)
plt.plot(x,sin,label = 'Sin')
plt.title('sin function')
plt.xlabel('X value')
plt.ylabel('Y value')

plt.subplot(1,2,2)
# x and cos(x)
plt.plot(x,cos,label = 'Cos',c = 'red')
plt.title('cos function')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```python
plt.figure(figsize=(15,5))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
ax1.plot(x,sin)
ax2.plot(x,cos,c='red')
ax1.set(xlabel = 'x value',ylabel = 'y value',title = 'sin function')
ax2.set(xlabel = 'x value',ylabel = 'y value',title = 'cos function')
plt.show()
```

```python

fig, ax = plt.subplots(1,2,figsize = (15,5))
ax[0].plot(x,sin)
ax[1].plot(x,cos,c = 'red')
plt.show()
```

### 2.2. Distribution/Histogram plot

```python
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
plt.hist(iris['petal_length'],bins = 20)
plt.show()
```

```python
iris['petal_length'].describe()
```

```python
sns.displot(iris['petal_length'],bins = 10);
```

```python
sns.histplot(iris['petal_length'],bins = 10);
```
