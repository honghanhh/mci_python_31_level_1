## 2. VISUALIZATION 

### 2.1. Line graph

#### 2.1.1 Plot 1 line graph

x ~ create an array range from -10 to 10 including 100 items
y ~ sin(x)

```python
x = np.linspace(-10,10,100)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.title('sin')

plt.show
```

**truc cos**

```python
x = np.linspace(-10,10,100)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.plot(x,sin ,label='Sin',c='red',linestyle ='dashdot')
plt.plot(x,cos, label='Cos',c='blue',linestyle='dotted')
plt.legend(loc='best')
plt.show()
```

```python
ax = plt.axes()
ax.plot(x,sin,label='Sin')
ax.plot(x,cos,label='Cos')
ax.set_title('Function')
ax.legend(loc='best')
# ax.set(xlabel='',ylabel='',title='') dien cac thong so cua do thi
plt.show()
```

#### 2.1.2 Plot 2 line graph

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
# x - sin x
# x - cos x
plt.plot(x,sin)
plt.title('Sin function')
plt.xlabel('X value')
plt.ylabel('y value ')
plt.subplot(1,2,2)
plt.plot(x,cos,c='green')
plt.title('Cos function')
plt.xlabel('x value')
plt.ylabel('Y value')
plt.show()
```

```python
plt.figure(figsize=(15,5))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
ax1.plot(x,sin)
ax2.plot(x,cos,c='red')
plt.show()
```

```python
fig ,ax=plt.subplots(1,2,figsize=(15,5))
ax[0].plot(x,sin)
ax[1].plot(x,cos,c='green',linestyle='dotted')
plt.show()
```

```python
sin_df = pd.DataFrame({'x':x,'sin':sin})
sin_df.head(5)
```

```python
sns.lineplot(data=sin_df,x='x',y='sin');
```

### 2.3. Scatter plot

```python
iris.head(10)
```

```python
iris.species.unique()
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width,label='sepal')
plt.scatter(iris.petal_length,iris.petal_width,label='petal')
plt.legend();
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',label='sepal')
sns.scatterplot(data=iris,x='petal_length',y='petal_width',label='petal')
plt.legend();
```

### 2.5. Pie chart

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
tips['sex'].value_counts()
```

```python
sex = tips.groupby('sex').total_bill.sum()
sex
```

```python
plt.pie(sex,labels=sex.index,autopct='%.2f%%');
```

```python
sex1 = sex.reset_index()
sex1
```

```python
sns.barplot(data=sex1,x='sex',y='total_bill');
```

```python
sex = tips.groupby('sex').total_bill.sum()
day = tips.groupby('day').total_bill.sum()
sex
```

```python
day
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].pie(sex,labels=sex.index, autopct='%.2f%%')
ax[1].pie(day,labels=day.index, autopct='%.2f%%');
```

```python
tips.head(2)
```

```python
male = tips[tips['sex']=='Male'].groupby('day').total_bill.sum()
male
```

```python
plt.pie(male, labels=male.index, autopct='%.2f%%',explode=[0,0.2,0,0]);
```

### 2.6. Swarmplot

```python
sns.swarmplot(data=tips, x ='time',y='total_bill');
```

### 2.7. Catplot

```python
sns.catplot(data=tips,x='time',y='total_bill',hue='smoker',kind='violin');
```

```python
sns.catplot(data=tips,x='time',y='total_bill',row='smoker',col='sex',hue='smoker',kind='violin');
```
