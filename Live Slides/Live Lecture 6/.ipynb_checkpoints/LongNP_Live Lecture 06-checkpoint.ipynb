### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo.info() #type object = strings
```

```python
type(chipo.item_price.iloc[0])
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
#Convert from string to float for item_price
chipo.item_price.iloc[0] #casting
```

```python
float('$2.39')
```

```python
'$2.39'[1:]
```

```python
def convert_type(x):
    """
    convert string type to float for value
    
    params:
    ------
    """
    return float(x[1:])
```

```python
chipo['item_price_float'] = chipo.item_price.apply(convert_type)
```

```python
def convert_type_2(x):
    """
    convert string type to float for value
    
    params:
    ------
    """
    return float(x.replace('$',''))
```

```python
chipo['item_price_float_2'] = chipo.item_price.apply(convert_type_2)
```

```python
# Lamda (Anonymous functions)
convert_type_3 = lambda x: float(x.replace('$',''))
```

```python
chipo['item_price_float_3'] = chipo.item_price.apply(convert_type_3)
```

## II. Filtering & Sorting

### Step 18. How many products cost more than $10.00?

```python
chipo['price_per_item'] = chipo['item_price_float_2'] / chipo['quantity'] #calculate price per unit
chipo.head()
```

```python
chipo['price_per_item'] = [x/y for x,y in list(zip(chipo['item_price_float_2'], chipo['quantity']))] 

#zip lấy từng phần từ của cột này tính toán với từng phần tử trong cột còn lại

chipo.head()
```

```python
chipo[(chipo['price_per_item'] > 10)]
```
