### 2.1. Line graph

#### 2.1.1 Plot 1 line graph

x - create an array range from -10 to 10 including 100 items

y - sin(x)

**Plot sin function given x**

```python
x = np.linspace(-10,10,100) #start = -10, end = 10, num=100
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.title('Sin Function')
plt.xlabel('X value')
plt.ylabel('Sin of x')
plt.grid()
plt.show()
```

**Plot sin and cos function given x**

```python
x = np.linspace(-10,10,100)
sin = np.sin(x)
cos = np.cos(x)
```

**Method 1**

```python
plt.plot(x, sin, label='Sin', c='green', linestyle='--')
plt.plot(x, cos, label='Cos', c='red')
plt.title('Function')
plt.xlabel('X value')
plt.ylabel('Function of x')
plt.legend(loc='best')
plt.show()
```

**Method 2**

```python
ax = plt.axes()
ax.plot(x,sin, label = 'Sin')
ax.plot(x,cos, label = 'Cos')
# ax.set_title('Function')
# ax.set_xlabel('X value')
# ax.set_ylabel('Function of x')
ax.set(xlabel='X value',ylabel ='Function of x', title='Function')
ax.legend(loc='best')
plt.show()
```

#### 2.1.2 Plot 2 line graph

**subplot** and **subplots**

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1) # 1- number of row, 2 - number of columns, 1-current position
# x and sin (x)
plt.plot(x, sin)
plt.title('Sin Function')
plt.xlabel('X value')
plt.ylabel('Y value')

plt.subplot(1,2,2)
# x and cos (x)
plt.plot(x,cos, c='red')
plt.title('Cos Function')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```python
plt.figure(figsize=(15,5))

ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax1.plot(x,sin)
ax2.plot(x,cos, c ='red')

ax1.set(xlabel='X', ylabel='Sin Function')
ax2.set(xlabel='X', ylabel='Cos Function')

plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(x,sin)
ax[1].plot(x,cos, c='red')
plt.show()
```

### 2.5. Pie chart

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
tips['sex'].value_counts()
```

**Male and Female distribution given total bills**

```python
sex = tips.groupby('sex').total_bill.sum()
sex
```

```python
type(sex)
```

```python
plt.pie(sex, labels=sex.index, autopct='%.2f%%');
```

```python
sex1 = sex.reset_index()
sex1
```

```python
sns.barplot(data=sex1, x='sex', y='total_bill'); #data x y
```

```python
sex = tips.groupby('sex').total_bill.sum()
day = tips.groupby('day').total_bill.sum()
```

```python
sex
```

```python
day
```

```python
#subplots
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].pie(sex, labels=sex.index , autopct='%.2f%%')
ax[1].pie(day, labels=day.index, autopct='%.2f%%');
```
