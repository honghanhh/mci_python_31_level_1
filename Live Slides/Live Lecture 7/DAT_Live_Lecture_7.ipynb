```python
import numpy as np 
```

```python
x = np.linspace(-10,10,100) #inclusivestart and inclusivestop
#It means that we creat an array which we 100 element
#np.arrange(0,100,step) #inclusive start and exclusive stop 
import numpy as np
y = np.arange(0,100,1)
y
```

```python
print(x)
```

```python
y = np.sin(x)
y
```

```python
#syntax
#1. plot what? plt.plot(x,y)
#2. details(title,xlabel,ylabel,legend,grid...)
plt.figure(figsize= (10,10))
plt.plot(x,y,c ='red')
plt.title('Sin function')
plt.xlabel('x_value')
plt.ylabel('y_value')
plt.grid('major')
plt.show()
```

**plot cos and sin function given x**

**subplot and subplots**

```python
plt.figure(figsize=(15,5))#Set the size of diagram for multiple value
plt.subplot(1,2,1)#check syntax
#(1,2,1) means that we have 1 row and 2 colums and 1 is the location of what diagram which we will plot it
# x and sinx here
plt.title("Sin Fuction")
plt.plot(x,sin,c='green',linestyle='--',label='Sin(x)')
plt.xlabel('x_value')
plt.ylabel('sin_value')
plt.legend(loc='best')
plt.grid('both')


plt.subplot(1,2,2)#check syntax
# x and cos(x) here
plt.title("Sin Fuction")
plt.plot(x,cos,c='red',label='Cos(x)')
plt.xlabel('x_value')
plt.xlabel('cos_value')
plt.legend(loc='best')

plt.show()
#WE just change on the third parameter
```

```python
plt.figure(figsize=(15,5))#Set the size of diagram for multiple value
ax1= plt.subplot(1,2,1)#check syntax
ax2= plt.subplot(1,2,2)#check syntax
# x and sinx here
ax1.plot(x,sin,c='red',label='Sin(x)')
ax1.set(xlabel='Xvalue',ylabel="Sin_Function",title = "Function of Sin")
ax1.legend(loc ='best')
ax1.grid('both')

#legend and whatever you can add more and more
ax2.plot(x,cos,c='green')
ax2.set(xlabel='Xvalue',ylabel="Cos_Function",title = "Function of Cos")
plt.show()
#WE just change on the third parameter
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,10))
ax[0].plot(x,sin,c='red',linestyle='--',label="Sin(x)")
ax[1].plot(x,cos,c='green')
ax[0].set(xlabel='x',ylabel='Sin(x)',title='SinFunction of x')
ax[1].set(xlabel='x',ylabel='Cos(x)',title='CosFunction of x')
ax[0].legend(loc='best')
plt.show()
```

```python
plt.figure(figsize = (7,7))
plt.boxplot(iris.sepal_length); #default:True
plt.title('The boxplot database of serial-sepal_length')
plt.show()
#box diagram(so so important at al) from min:4.3 to max:7.9 mean =5.8 
```

```python
plt.boxplot(iris.sepal_length,vert=False); #rotate the box plot as y axis
```

***Seabors'sApplicationb for showing data frame***

- Note that seaborn's syntax: data = what? x=what(horizontal) and y=what(vertical)

```python
plt.figure(figsize = (5,5))
sns.boxplot(data = iris, x ='sepal_length',width = 0.2,color = 'red');
plt.title('The box plot following X"axis')
```

### 2.5. Pie chart

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['sex'].value_counts().sum()
```

```python
#Male and Female...
#Totalbill
```

```python
tips.groupby('sex')['total_bill'].sum()
```

```python
tips.groupby('sex').total_bill.sum()
```

```python
sex = tips.groupby('sex').total_bill.sum()
sex
```

```python
plt.pie(sex);
```

```python
plt.pie(sex, labels = sex.index);
```

```python
plt.pie(sex, labels = sex.index,autopct='%.2f%%');#means that automatically calculate the sex
```

```python
sex1 = sex.reset_index()
sex1
```

```python
# We have more than one
```

```python
sns.barplot(data= sex1, x='sex',y= 'total_bill' )
```

```python
sns.barplot(data= sex1, x='sex',y= 'total_bill' )
```

```python
sex = tips.groupby('sex').total_bill.sum()
day = tips.groupby('day').total_bill.sum()
sex
```

```python
day
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,5))
```
