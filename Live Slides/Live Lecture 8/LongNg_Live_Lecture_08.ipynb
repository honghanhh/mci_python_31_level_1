### 4.4 Model

Model should have
- Train samples
- Test samples
ratio: 7/3, 8/2, 8.5/1.5

#### 4.4.1 Train test plit

```python
iris.columns
```

```python
X_iris = [['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
X_iris = iris.drop('species', axis = 1)
```

```python
y_iris = iris['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, test_size = 0.2, random_state = 42)
```

```python
print(Xtrain.shape)
print(Xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
print(Xtrain.shape)
print(Xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

#### 4.4.2 Train model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
#model1 = LogisticRegression(max_iter = 1000)
model1 = SVC()
```

```python
model1.fit(Xtrain, ytrain)
```

#### 4.4.2 Predict model

```python
y_pred = model1.predict(Xtest)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(ytest,y_pred)
```
