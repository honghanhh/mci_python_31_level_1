**Plot sin and cos function given x**

```python
x = np.linspace(-10, 10, 100)
sin = np.sin(x)
cos = np.cos(x)
plt.plot(x, sin, label='sin', c='green', linestyle='--')
plt.plot(x, cos, label='cos', c='red', linestyle='-.')
plt.legend(loc='best')
plt.show()
```

```python
ax = plt.axes()
ax.plot(x,sin)
ax.plot(x,cos)
ax.set(xlabel='X value', ylabel='Function of x', title='Function')
plt.show()
```

**2. subplot and subplots**

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1) # 1 - number of row, 2 - number of columns, 1 - current position
plt.title('Sin funtion')
plt.plot(x, sin)
plt.xlabel('X value')
plt.ylabel('Y value')

plt.subplot(1, 2, 2)
plt.title('Cos funtion')
plt.plot(x, cos, c='red')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```python
plt.figure(figsize=(15,5))
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1,2,2)

ax1.plot(x, sin)
ax2.plot(x, cos, c='red')
ax1.set(xlabel='X', ylabel='Y', title='Sin')
ax2.set(xlabel='X', ylabel='Y', title='Cos');
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(x, sin)
ax[1].plot(x, cos, c='red');
```

**2.5 pie chart**

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
tips['sex'].value_counts()
```

**Male and Female distribution given total bills**

```python
sex= tips.groupby('sex').total_bill.sum()
sex
```

```python
plt.pie(sex, labels=sex.index, autopct='%.2f%%');
```

```python
sex1 = sex.reset_index()
sex1
```

```python
sns.barplot(data=sex1, x='sex', y='total_bill')
```

```python
sex= tips.groupby('sex').total_bill.sum()
day = tips.groupby('day').total_bill.sum()
sex
```

```python
day
```

```python
#subplots
fig, ax = plt.subplots(1,2, figsize=(15, 5))
ax[0].pie(sex, labels=sex.index, autopct='%.2f%%')
ax[1].pie(day, labels=day.index, autopct='%.2f%%');
```

**2.6 swarmplot**

```python
sns.swarmplot(data=tips, x='time', y='total_bill', hue='smoker');
```
