## 2. VISUALIZATION 

### 2.1. Line graph

#### 2.1.1 Plot 1 line graph

x - create an array range from -10 to 10 incluing 100 items
y - sin(x)

```python
x = np.linspace(-10,10,100)  # bat dau tu 10, ket thuc la 10 vaf 100 phan tu
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.title('Sin funtion') # Them ten do thi
plt.xlabel('x value') # truc sin
plt.ylabel('sin of x') # truc cos
plt.grid() # o luoi
plt.show() # xoa mo ta kieu du lieu
```

```python
sin_df = pd.DataFrame({'x':x, 'y':y})
sin_df.head(5)
```

```python
sns.lineplot(data=sin_df, x='x', y='y')
plt.show()
```

## Plot sin and cos function given x

```python
x = np.linspace(-10,10,100)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.plot(x, sin, label='sin', c='green', linestyle='--') # line style de chon lai duong ve~
plt.plot(x, cos, label='cos', c='red')
plt.title('Function')
plt.xlabel('X value')
plt.ylabel('Function of x')
#plt.legend(loc='lower left')
plt.legend(loc='best') # chon vi tri tot nhat mo ta ten cot, /muon sd plt.legen thi phai them label/
plt.show()
```

Method 2

```python
ax = plt.axes()
ax.plot(x,sin, label='Sin')
ax.plot(x,cos, label='Cos')
#ax.set_title('Function')
#ax.set_xlabel('x value')
#ax.set_ylabel('Function of x')
ax.set(xlabel='X value', ylabel='Function of x', title='Fuction')
ax.legend(loc='best')
plt.show()
```

#### 2.1.2 Plot 2 line graph

**subplot** and subplots

```python
plt.figure(figsize= (15,5))
plt.subplot(1,2,1)
#x and sin x
plt.plot(x, sin)
plt.title('Sin Funtion')
plt.xlabel('X value')
plt.ylabel('Y value')

plt.subplot(1,2,2) # tham so chi thay doi o vi tri thu 3 de thay doi vi tri
#
plt.plot(x,cos, c='red')
plt.title('Cos Function')
plt.xlabel('X value')
plt.ylabel('Y value')

plt.show()
```

```python
plt.figure(figsize=(15,5))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax1.plot(x,sin)
ax2.plot(x, cos, c='red')

ax1.set(xlabel='X', ylabel='sin Function')
ax2.set(xlabel='X', ylabel='cos Function')

plt.show()
```

### 2.5. Pie chart

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['sex'].value_counts()
```

```python
# Male and Female distribution given total bill /Tinh so luong % gioi tinh nam, nu trong tap DL/
```

```python
sex = tips.groupby('sex').total_bill.sum()
sex
```

```python
plt.pie(sex, labels=sex.index, autopct='%.2f%%'); ## them label de them phan mo  ta vao bieu do
```

```python
#sns khong the bieu dien pie chart duoc
```

```python
sex1 = sex.reset_index()
sex1
```

```python
sns.barplot(data=sex1, x='sex', y='total_bill');
```

```python
sex = tips.groupby('sex').total_bill.sum()
day = tips.groupby('day').total_bill.sum()
```

```python
sex
```

```python
day
```

```python
#subplots
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].pie(sex, labels=sex.index, autopct='%2.f%%');
ax[1].pie(day, labels=day.index, autopct='%.2f%%');
```
